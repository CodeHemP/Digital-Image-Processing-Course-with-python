```
# Importar librerias 

import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline  

plt.rcParams['figure.figsize'] = [10,5]
```

```
# Leer una imagen
imgbgr = cv2.imread('arcoiris.jpg')
print('Tamaño :',imgbgr.shape)
```

```
# Ploteando imagen con matplotlib
imgrgb = cv2.cvtColor(imgbgr, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
plt.show()
```

```
# Ploteando los 3 canales
plt.figure()
plt.subplot(131)
plt.imshow(imgrgb[:,:,0])
plt.title('R channel')
plt.subplot(132)
plt.imshow(imgrgb[:,:,1])
plt.title('G channel')
plt.subplot(133)
plt.imshow(imgrgb[:,:,2])
plt.title('B channel')
plt.show()
```
