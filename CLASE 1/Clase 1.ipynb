```python
# Ploteando los 3 canales
# IMGRGB 

plt.figure() # CREA EL ESPACIO DE UNA IMAGEN

plt.subplot(131)
plt.imshow(imgrgb[:,:,0],cmap='gray')
plt.title('R channel')
plt.subplot(132)
plt.imshow(imgrgb[:,:,1],cmap='gray')
plt.title('G channel')
plt.subplot(133)
plt.imshow(imgrgb[:,:,2],cmap='gray')
plt.title('B channel')

plt.show()  # MUESTRA LA IMAGEN DE LA FIGURA
```
